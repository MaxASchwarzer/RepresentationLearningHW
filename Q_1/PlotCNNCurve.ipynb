```python
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("cnn_trace.pkl", "rb") as f:
    train_accs, val_accs, train_losses, val_losses = pkl.load(f)
```

```python
fig, ax1 = plt.subplots(figsize=(5, 5))
ax2 = ax1.twinx()
ln1 = ax1.plot(np.arange(1, len(train_accs)+1), train_accs, c="red", label="Train Accuracy")
ln2 = ax1.plot(np.arange(1, len(train_accs)+1), val_accs,  linestyle=":", c="red", label="Valid Accuracy")

ln3 = ax2.plot(np.arange(1, len(train_losses)+1),  train_losses, c="black", label="Train Loss")
ln4 = ax2.plot(np.arange(1, len(val_losses)+1), val_losses,  linestyle=":", c="black", label="Valid Loss")

lns = ln1+ln2+ln3+ln4
labels = [ln.get_label() for ln in lns]

ax1.set_ylabel("Accuracy")
ax2.set_ylabel("Loss")
ax1.set_xlabel("Epoch")
ax1.legend(lns, labels, loc="right")
plt.savefig("CNN_training_curve.png", dpi=200)
```
